```python
import plotly.express as px
fig = px.bar(x=['a', 'b', 'c'], y = [1, 3, 2])
fig.write_html('first_figure.html', auto_open=True)
```

```python
fig.show()
```

```python
import numpy as np
fig = px.scatter(x=np.arange(10), y=np.arange(10)**2)
fig.show()
```

```python
fig = px.sc atter(x=np.random.randn(10_000), y=np.random.randn(10_000))
fig.show()
```

```python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", size="petal_length", hover_data=['petal_width'])
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="petal_length", size="petal_length", hover_data=['petal_width'])
fig.show()
```

```python
fig = px.scatter(df,x="sepal_width", y="sepal_length", color="species", symbol="species")
fig.show()
```
